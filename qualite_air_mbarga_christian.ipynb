```python
data2 = data1[data1["date_ech"] < "2020-03-17T10:00:00"]
data3 = data1[data1["date_ech"] > "2020-03-18T10:00:00"][data1["date_ech"] < "2020-05-16T10:00:00"]
```

```python
data2 = data2.groupby("lib_zone").mean()
data3 = data3.groupby("lib_zone").mean()
```

```python
names = ["Brest", "Vannes", "Lorient", "Quimper", "Saint-Brieuc", "Saint-Malo", "Rennes"]
coord = [data1["longitude"].min()-0.7, data1["longitude"].max()+0.8, data1["latitude"].min()-0.4, data1["latitude"].max()+0.3]
```

```python
plt.imshow(plt.imread("bretagne74.gif"), extent=coord)

scatter1 = plt.scatter(data2["longitude"], data2["latitude"], c=data2["valeur"], s=200, vmin=3, vmax=5)
scatter1

for i, txt in enumerate(data2["valeur"]):
    plt.annotate((names[i], round(txt, 2)), (data2["longitude"][i], data2["latitude"][i]))

plt.title("Moyennes de l'indice de pollution, 1 janvier-16 mars 2020")
fig = plt.gcf()
fig.set_size_inches(15, 5)
cbar1 = fig.colorbar(scatter1, ticks=[3, 4, 5])
cbar1.ax.set_yticklabels(["Mauvais", "Moyen", "Bon"])
plt.show()
```

```python
plt.imshow(plt.imread("bretagne74.gif"), extent=coord)

scatter2= plt.scatter(data3["longitude"], data3["latitude"], c=data3["valeur"], s=200, vmin=3, vmax=5)
scatter2

for i, txt in enumerate(data3["valeur"]):
    plt.annotate((names[i], round(txt, 2)), (data3["longitude"][i], data3["latitude"][i]))

plt.title("Moyennes de l'indice de pollution, 19 mars-15 mai 2020")
fig = plt.gcf()
fig.set_size_inches(15, 5)
cbar2 = fig.colorbar(scatter2, ticks=[3, 4, 5])
cbar2.ax.set_yticklabels(["Mauvais", "Moyen", "Bon"])
plt.show()
```
